```python
## Criando a classe do data preparation
class WineQuality(object):
    def __init__(self):
        self.free_sulfur_scaler = pickle.load(open('free_sulfur_scaler.pkl', 'rb'))
        self.total_sulfur_scaler = pickle.load(open('total_sulfur_scaler.pkl', 'rb'))
    
    def data_preparation(self, df):
        ## reescalando o sulfur 
        df['free sulfur dioxide'] = self.free_sulfur_scaler.transform(df[['free sulfur dioxide']])
        df['total sulfur dioxide'] = self.total_sulfur_scaler.transform(df[['total sulfur dioxide']])

        return df
```
